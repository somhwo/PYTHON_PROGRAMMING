```python
import numpy as np
import pandas as pd
from scipy import stats
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
import torch
import matplotlib.pyplot as plt
```

```python
# Generate synthetic data using sklearn's make_blobs
points, labels = make_blobs(
    n_samples=200,
    centers=3,
    n_features=2,
    random_state=38
)

# Pandas
table = pd.DataFrame({
    "X value": points[:, 0],
    "Y value": points[:, 1],
    "Actual Group": labels
})

table.head()
```

```python
# Initialize & fit K-means
model = KMeans(n_clusters=3, random_state=42)
predicted = model.fit_predict(table[["X value", "Y value"]])

table["Group Found"] = predicted
table.head()
```

```python
# Visualize using matplotlib
plt.figure(figsize=(7, 5))

plt.scatter(
    table["X value"],
    table["Y value"],
    c=table["Group Found"],
    s=40,
    alpha=0.8
)

# plot cluster centers
centers = model.cluster_centers_
plt.scatter(
    centers[:, 0],
    centers[:, 1],
    s=150,
    marker="X",
    color="red"
)

plt.xlabel("X value")
plt.ylabel("Y value")
plt.grid(True)

plt.show()
```
